```python
file_name = "rowsReduced.csv"
rows=pd.read_csv(file_name,delimiter=',') # Load and display the data as a dataframe
rows.head()
```

```python
del rows['Unnamed: 0']
```

# Adding the name of the genes

```python
names=[None]*1351
```

```python
for i in range(0,1351):
    names[i]=rows['gene_symbol'][i]
```

```python
names
```

```python
matr2['gene_symbol']=names
```

```python
nameColumn = matr2['gene_symbol']
matr2.drop(labels=['gene_symbol'], axis=1,inplace = True)
matr2.insert(0, 'gene_symbol', nameColumn)
matr2.head()
matr2.to_csv('matrixRaw.csv')
```
