```python
# As always
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mpl
from scipy.interpolate import *
mpl.rcParams.update({'font.size':24})


labels=['DFC','HIP']
file_name = "dataMatrix.xlsx"
DFC=pd.read_excel(file_name,sheet_name='DFC') # Load and display the data as a dataframe
file_name = "dataMatrix.xlsx"
HIP=pd.read_excel(file_name,sheet_name='HIP') # Load and display the data as a dataframe
```

```python
DFC.reset_index(inplace  =False)
```

```python
HIP.reset_index(inplace  =False)
```

```python
DFC=DFC.reset_index(drop =True)
HIP=HIP.reset_index(drop =True)
```

```python
indexToName={51:'VIM'
,83:'NNAT'
,147:'CRMP1'
,298:'STMN2'
,368:'YWHAE'
,416:'DPYSL3'
,462:'STMN1'
,517:'SOX4'
,679:'TUBB2B'
,815:'GPM6A'
,850:'UCHL1'
,1003:'EEF2'
,1054:'UBB'
,592:'MAP1B'}
nameToColor={'VIM':0
,'NNAT':1
,'CRMP1':2   
,'STMN2':3
,'YWHAE':4
,'DPYSL3':5
,'STMN1':6
,'SOX4':7
,'TUBB2B':8
,'GPM6A':9
,'UCHL1':10
,'EEF2':11
,'UBB':12
,'MAP1B':13}
nameToColor[indexToName[51]]

lineMarkers=[x+y for x in ['b','g','r','c','m','y','k'] for y in ['--','-']]
coordinateMarkers=[x+y for x in ['b','g','r','c','m','y','k'] for y in ['o','>']]
x_i=np.linspace(8,24,100)
```

```python
#changed for DFC
x_given=[8,9,12,13,16,17,19,21,24,26,37]
xindex=[0,1,2,5,8,11,12,13,14,15,16]



fig=plt.figure()

for changable in range(0,1351):
    y_given=DFC.iloc[xindex,changable].tolist()
    f_spline=interp1d(x_given,y_given, kind='cubic')
    y_is=f_spline(x_i)
    
    if min(y_is) < 0:
        z=1
    elif max(y_is) > 500:
        plt.plot(x_i,y_is,lineMarkers[nameToColor[indexToName[changable]]])
        #print(changable)
        plt.xlim(8,24)
    else:
        z=1

for changable in range(0,1351):
    y_given=DFC.iloc[xindex,changable].tolist()
    f_spline=interp1d(x_given,y_given, kind='cubic')
    y_is=f_spline(x_i)
    
    if min(y_is) < 0:
        z=1
    elif max(y_is) > 500:
        plt.plot(x_given,y_given,coordinateMarkers[nameToColor[indexToName[changable]]])
        #print(changable)
        plt.xlim(8,24)
        
    else:
        z=1
       
    
plt.ylim(0,3000)
fig.set_size_inches(18.5, 10.5)
plt.ylabel('Gene Expression Level')   
plt.xlabel('Weeks')  


plt.title('DFC')
plt.legend(['VIM', 'NNAT', 'CRMP1', 'STMN2', 'YWHAE', 'DPYSL3', 'STMN1', 'SOX4',
       'TUBB2B', 'GPM6A', 'UCHL1', 'EEF2', 'UBB'],loc=2)
plt.savefig('MajorDFC')
```
